```python
import pylab as plt 
import numpy as np 
import netCDF4 as nc 


```

```python
f = 'w_scratch.nc'
d = nc.Dataset(f,'r')
data = d.variables['data']
x = d.variables['xh'][:]
y = d.variables['yh'][:]

print d.variables.keys()
```

```python
plt.figure(figsize=(14,10)) 
plt.contour(data[:,:,25].T, 100)
plt.colorbar()
plt.show() 
```

```python
from scipy import interpolate
```

```python
f = interpolate.interp2d(x , y  , data[:,:,25], kind='cubic', bounds_error=True)
f_lienar =  interpolate.interp2d(x , y  , data[:,:,25], kind='linear', bounds_error=True)
```

```python
x2 = np.linspace(x.min(), x.max(), 1024)
y2 = np.linspace(y.min(), y.max(), 1024)


d2 = f(x2,y2)
dl = f_lienar(x2,y2)
```
